```python
import math

def is_prime(num_34):
    """Fungsi untuk memeriksa apakah sebuah bilangan adalah prima."""
    if num_34 <= 1:
        return False
    if num_34 <= 3:
        return True
    if num_34 % 2 == 0 or num_34 % 3 == 0: # arti dari % itu adalah mod
        return False
    i = 5
    while i * i <= num_34:
        if num_34 % i == 0 or num_34 % (i + 2) == 0: # arti dari % itu adalah mod
            return False
        i += 6
    return True

# Tentukan nilai n
n_34 = 34

# menghitung nilai 
a_34 = 1000 * n_34
b_34 = 1000 * (n_34 + 1)

# Cari bilangan prima dalam rentang 
bilangan_prima_34 = []
for num_34 in range(a_34, b_34):
    if is_prime(num_34):
        bilangan_prima_34.append(num_34)

# Cetak bilangan prima yang ditemukan
print(f"Bilangan prima antara {a_34} dan {b_34} adalah:")
print(bilangan_prima_34)
```
