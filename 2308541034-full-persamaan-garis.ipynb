```python
import numpy as np
import matplotlib.pyplot as plt
#Fungsi untuk menghitung koefisien persamaan garis dari dua titik
def line_coefficients(x1,y1,x2,y2):
    a_34 = y2-y1
    b_34 = x1-x2
    c_34 = a_34*x1+b_34*y1
    return a_34,b_34,c_34
 
#Fungsi unntuk menemukan titik potong dari dua garis
def find_intersection(a1_34,b1_34,c1_34,a2_34,b2_34,c2_34):
    A = np.array([[a1_34,b1_34],[a2_34,b2_34]])
    C = np.array([c1_34,c2_34])
    if np.linalg.det(A) == 0:
        return None #garis garis sejajar atau berimpit
    else:
        intersection = np.linalg.solve(A,C)
        return intersection
# Titik -titik untuk dua garis
x1_34 = 1
y1_34 = 2
x2_34 = 3
y2_34 = 4
x3_34 = 2
y3_34 = 3
x4_34 = 4
y4_34 = 1
#hitung koefisien garis pertama
a1_34,b1_34,c1_34 = line_coefficients(x1_34,y1_34,x2_34,y2_34)
#hitung koefisien garis kedua
a2_34,b2_34,c2_34 = line_coefficients(x3_34,y3_34,x4_34,y4_34)
#cari titik potong
intersection = find_intersection(a1_34,b1_34,c1_34,a2_34,b2_34,c2_34)
#buat plot garis dan titik potong
fig, ax = plt.subplots()
#Garis pertama 
x_vals = np.linspace(0,5,400)
y_vals1 =(c1_34-a1_34*x_vals)/ b1_34
ax.plot(x_vals,y_vals1,label="Garis 1")
#Garis kedua
y_vals2 = (c2_34-a2_34*x_vals)/ b2_34
ax.plot(x_vals,y_vals2,label="Garis 2")
#Titik potong
if intersection is not None:
    ax.plot(intersection[0]),intersection[1],('ro') #'ro' berarti titik merah
    ax.annotate(f'Intersection\n({intersection[0]:.2f}):,{intersection[1]:.2f})',
                xy=(intersection[0],intersection[1]),xytext=(intersection[0]+0.5,intersection[1]-0.5),
                arrowprops=dict(facecolor='black',shrink=0.05))
# Plot titik titk yang digunakan untuk membuat garis
ax.plot([x1_34,x2_34],[y1_34,y2_34],'bo')# Titik biru untuk garis 1
ax.plot ([x3_34,x4_34],[y3_34,y4_34],'go')# Titik hijau untuk garis 2
#tambahkan label dan legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)
#Tampilkan plot
plt.show()import numpy as np
import matplotlib.pyplot as plt
#Fungsi untuk menghitung koefisien persamaan garis dari dua titik
def line_coefficients(x1,y1,x2,y2):
    a_34 = y2-y1
    b_34 = x1-x2
    c_34 = a_34*x1+b_34*y1
    return a_34,b_34,c_34
 
#Fungsi unntuk menemukan titik potong dari dua garis
def find_intersection(a1_34,b1_34,c1_34,a2_34,b2_34,c2_34):
    A = np.array([[a1_34,b1_34],[a2_34,b2_34]])
    C = np.array([c1_34,c2_34])
    if np.linalg.det(A) == 0:
        return None #garis garis sejajar atau berimpit
    else:
        intersection = np.linalg.solve(A,C)
        return intersection
# Titik -titik untuk dua garis
x1_34 = 1
y1_34 = 2
x2_34 = 3
y2_34 = 4
x3_34 = 2
y3_34 = 3
x4_34 = 4
y4_34 = 1
#hitung koefisien garis pertama
a1_34,b1_34,c1_34 = line_coefficients(x1_34,y1_34,x2_34,y2_34)
#hitung koefisien garis kedua
a2_34,b2_34,c2_34 = line_coefficients(x3_34,y3_34,x4_34,y4_34)
#cari titik potong
intersection = find_intersection(a1_34,b1_34,c1_34,a2_34,b2_34,c2_34)
#buat plot garis dan titik potong
fig, ax = plt.subplots()
#Garis pertama 
x_vals = np.linspace(0,5,400)
y_vals1 =(c1_34-a1_34*x_vals)/ b1_34
ax.plot(x_vals,y_vals1,label="Garis 1")
#Garis kedua
y_vals2 = (c2_34-a2_34*x_vals)/ b2_34
ax.plot(x_vals,y_vals2,label="Garis 2")
#Titik potong
if intersection is not None:
    ax.plot(intersection[0]),intersection[1],('ro') #'ro' berarti titik merah
    ax.annotate(f'Intersection\n({intersection[0]:.2f}):,{intersection[1]:.2f})',
                xy=(intersection[0],intersection[1]),xytext=(intersection[0]+0.5,intersection[1]-0.5),
                arrowprops=dict(facecolor='black',shrink=0.05))
# Plot titik titk yang digunakan untuk membuat garis
ax.plot([x1_34,x2_34],[y1_34,y2_34],'bo')# Titik biru untuk garis 1
ax.plot ([x3_34,x4_34],[y3_34,y4_34],'go')# Titik hijau untuk garis 2
#tambahkan label dan legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)
#Tampilkan plot
plt.show()
```
