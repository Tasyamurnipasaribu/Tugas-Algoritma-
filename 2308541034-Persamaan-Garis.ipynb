```python
import numpy as np 
import matplotlib.pyplot as plt 

#persamaan garis pertama melalui titik (x1,y1) dan (x2,y2)
x1 = 2
y1 = 3
x2 = 4
y2 = 7
#mencari gradien
m1_34=(y2-y1)/(x2-x1)
print("Maka dapat kita ketahui bahwa gradiennya adalah {}".format(m1_34))
print("Persamaan garis pertama adalah y={}x+{}".format(m1_34,-m1_34*x1+y1))
#y=m1_34*(x-x1)+y1

#persamaan garis kedua melalui titik (x3,y3) dan (x4,y4)
x3 = 2
y3 = 3
x4 = 4
y4 = -7
#mencari gradien
m2_34 =(y4-y3)/(x4-x3)
print("Maka dapat kita ketahui bahwa gradiennya adalah {}".format(m2_34))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_34,-m2_34*x3+y3))
#y=m2_34*(x-x3)+y3
if m1_34==m2_34:
    print("karena Gradiennya sama, maka kemungkinan tidak ada solusi atau bisa disebut solusi nya banyak")

else:
    # Mencari titik potong pada kedua garis yaitu (tp_x,tp_y)
    tp_x = (m1_34*x1-y1*x3+y3)/(m1_34-m2_34)
    tp_y = (m2_34*(tp_x-x3)+y3)
    
    print("Titik potong adalah ({},{})".format(tp_x,tp_y))
    x=np.arange(tp_x-5,tp_x+5,0.01)
    plt.figure(figsize=(4,4),dpi=200)
    
    plt.plot(x,m1_34*(x-x1)+y1,color='yellow',label='y={}x+{}'.format(m1_34,-m1_34*x1+y1))
    plt.plot(x,m2_34*(x-x2)+y2,color='pink',label='y={}x+{}'.format(m2_34,-m2_34*x3+y3))
    plt.plot(tp_x,tp_y,color='green',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Persamaan garis pertama melalui titik (a,b) dan (c,d)
a = 5
b = 6
c = 6
d = 7
# Persamaan garis kedua melalui titik (e,f) dan (g,h)
e = 8
f = 8
g = 10
h = 11
m1_34 = (d-b)/(c-a)
m2_34 = (h-f)/(g-e)
print("maka dapat kita ketahui gradien garis pertama adalah {}".format(m1_34))
print("maka dapat kita ketahui gradien garis kedua adalah {}".format(m2_34))
x = np.arange(-3,3,0.02)
plt.plot(x,m1_34*(x-a)+b,color='red')
plt.plot(x,m2_34*(x-e)+f,color='purple')
plt.show()
```
