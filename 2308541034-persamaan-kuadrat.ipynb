```python
# Menentukan persamaan kuadrat jika diketahui akar akarnya 
# Menentukan grafik fungsi kuadrat
# Menentukan titik puncak dan sumbu simetri
# Menentukan posisi titik puncak (dikiri sumbu y,pada sumbu y, di kanan sumbu y)
# Bagaimana kondisi parabola memotong sumbu y (potong sumbu y positif, potong sumbu y di pusat(0,0),potong sumbu y negatif)
# Macam macam akar persamaan kuadrat dilihat dari diskriminan
# Penyelesaian persamaan kuadrat dapat menggunakan rumus ABC
# Menyusun persamaan kuadrat baru jika diketahui akar akarnya
import numpy as np
import matplotlib.pyplot as plt
import cmath
# saya akan menentukan  nilai a,b, c
a_34 = float(input("nilai dari a adalah  ")) # a != 0
b_34 = float(input("nilai dari b adalah "))
c_34 = float(input("nilai dari c adalah "))
print("Persamaan Kuadratnya adalah ({}x**2+{}x+{})".format(a_34,b_34,c_34))

# menghitung diskriminan 
D_34 = b_34**2-4*a_34*c_34

# Menentukan jenis akar berdasarkan nilai diskriminan
if D_34 > 0:
     print("Persamaan memiliki dua akar riil yang berbeda.")
elif D_34 == 0:
     print( "Persamaan memiliki satu akar riil (akar kembar).")
else:
     print("Persamaan memiliki dua akar imajiner.")
        
# Menghitung akar-akarnya
akar1_34 = (-b_34 - cmath.sqrt(D_34)) / (2 * a_34)
akar2_34 = (-b_34 + cmath.sqrt(D_34)) / (2 * a_34)
print(f"Akar-akarnya adalah ({akar1_34}, {akar2_34})")

#menentukan titik puncak dan sumbu simetri
h_34 = -b_34/ (2*a_34)
k_34 = a_34*h_34**2 + b_34*h_34 +c_34
print(f"nilai h adalah ({h_34})")
print(f"nilai k adalah({k_34})")  

# Menentukan posisi puncak dan sumbu simetri y 
if h_34 < 0:
    print("titik puncak berada dikiri sumbu y.")
elif h_34==0:
    print("titik puncak berada tepat pada sumbu y.")
else:
    print("titik puncak berada dikanan sumbu y.")
    
# Mentukan kondisi memotong sumbu y 
if c_34 > 0:
    print("Parabola memotong sumbu y secara positif.")
elif c_34 ==0:
    print("Parabola memotong sumbu y tepat pada pusat (0,0).")
else:
    print("Parabola memotong sumbu y secara negatif.")

# Menentukan koefisien dari persamaan kuadrat baru
d_34 = 1
e_34 = -2 * (akar1_34 + akar2_34 )
f_34 = akar1_34  * akar1_34 
print("Persamaan Kuadratnya adalah ({}x**2+{}x+{} =0)".format(d_34,e_34,f_34))

    
# Plot kurva persamaan kuadrat
x_values = np.linspace(-10, 10, 400)
y_values = a_34 * x_values**2 + b_34 * x_values + c_34
plt.plot(x_values, y_values, label=f'{a_34}x^2 + {b_34}x + {c_34}')
plt.plot ("persamaan kuadrat")
plt.plot ("persamaan kuadrat baru")


# tandai titik puncak
plt.plot(h_34,k_34,'ro', label =f"Titik puncak:({h_34},{k_34})")

# tandai sumbu simetri 
plt.axvline(x=h_34,color ='green',linestyle='--',label=f"Sumbu simetri:x={h_34}")

# Menambahkan garis sumbu dan titik-titik akar
plt.axhline(0, color='pink', linewidth=0.5)
plt.axvline(0, color='pink', linewidth=0.5)
plt.grid(color='green', linestyle='--', linewidth=0.5)

# Menandai akar-akarnya jika real
if D_34 >= 0:
    plt.scatter([akar1_34.real, akar2_34.real], [0, 0], color='red', zorder=5)

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafik Persamaan Kuadrat dan kuadrat baru')
plt.legend()
plt.grid(True)
plt.show()
```
